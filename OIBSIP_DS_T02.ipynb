```python
feature=[' Estimated Unemployment Rate (%)',' Estimated Employed',' Estimated Labour Participation Rate (%)']
sn.pairplot(df[feature])
plt.show()
```

```python
df.columns=['Region','Date','Frequency','Estimated Unemployment Rate',
                'Estimated Employed','Estimated Labour Participation Rate',
                'Area']
plt.title('Indian Unemployment')
sn.histplot(x='Estimated Employed',hue='Area',data=df)
plt.show()
```

```python
group_state=df.groupby(['Region'])[['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate']].mean().reset_index()
```

```python
group_state
```

```python
group_area=df.groupby(['Area'])[['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate']].mean().reset_index()
group_area
```

```python
df1=df.groupby('Region')[['Estimated Unemployment Rate']].mean().reset_index()
sn.barplot(x='Estimated Unemployment Rate',y='Region',data=df1)
plt.title('Bar plot of Estimated Unemployment Rate according to State')
plt.show()
```

```python
df2=df.groupby('Region')['Estimated Employed'].mean().reset_index()
sn.barplot(x='Estimated Employed',y='Region',data=df2)
plt.title('Bar plot of Estimated Employment according to State')
plt.show()
```

```python
df3=df.groupby('Region')['Estimated Labour Participation Rate'].mean().reset_index()
sn.barplot(x='Estimated Labour Participation Rate',y='Region',data=df3)
plt.title('Bar plot of Estimated Labour Employment Rate according to State')
plt.show()
```
